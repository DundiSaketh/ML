```
import numpy as np
import matplotlib.pyplot as plt
x= np.array([1,2,3])
y=np.array([1.2,1.9,3.2])
a=x.sum()
b=y.sum()
c=(x*x).sum()
d=(y*y).sum()
e=len(x)
m1=(c*b-a*d)/(c*e-a*a) # m1 is intercept
m2=(b-(e*m1))/a # m2 is slope
values=[m2*i+ m1 for i in x]
plt.scatter(x,y,)
plt.plot(x,values,'m')
plt.show() 


```
